```python
import pandas as pd
```

```python
with open('company_policy.txt', 'r', encoding='utf-8') as file:
    raw_text = file.read()
```

```python
print(raw_text[:500])  # Print the first 500 characters to verify content
```

```python
len(raw_text)
```

```python
lines = raw_text.split('\n')
len(lines)
```

```python
for line in range(10):
    print(f"{line + 1}.{lines[line]}")
```

```python
clean_lines = []
```

```python
for line in lines:
    line=line.strip()
    if len(line) > 0:
        clean_lines.append(line)

clean_lines[:10]
```

```python
len(clean_lines)
```

```python
word_counts = []
```

```python
for line in clean_lines:
    word_counts.append(len(line.split()))

print(word_counts[:10])
```
